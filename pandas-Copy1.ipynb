Plot กราฟ lattitude longitude day

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
axe = Axes3D(fig)
axe.scatter(data7['LATITUDE'], data7['LONGITUDE'], data7['diffdaymin'])
axe.set_xlabel('lat(X)')
axe.set_ylabel('lon(Y)')
axe.set_zlabel('day(Z)')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN
```

```python
#เก็บค่าlatitude 7 วันไว้ในตัวแปร rray
rray = np.array(data7.values.tolist())
rray
```

```python
#สร้าง model ในการจำแนกข้อมูล
model = DBSCAN(eps=5, min_samples=25)
model.fit_predict(data7)
pred = model.fit_predict(data7)

fig = plt.figure()
axes = Axes3D(fig)
axes.scatter(rray[:,0], rray[:,1], rray[:,2], c = model.labels_, s=50)
axes.view_init(azim=200)
plt.show()
```

```python
model_label = model.labels_
```

```python
#Calculate silhouette 
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_score(rray, model_label )
```

ค่า silhouette ที่คำนวณได้มีค่าคือ 0.9980917942362771 โดยได้กำหนดค่า eps = 5 เพราะต้องการให้ข้อมูลที่ได้มีการกระจายตัวในพื้นที่กว้าง โดยได้ใช้เทคนิคDensity Based Spatial Clustering of Applications with Noise ซึ่งจากการคำนวณของโปรแกรมแล้วได้ค่า silhouette เท่ากับ 0.9980917942362771 แสดงว่าช่วงข้อมูลที่เลือกมากับวิธีคำนวณที่ใช้มีความแม่นยำมากและมีค่า outlier ไม่มากจนเกินไป
